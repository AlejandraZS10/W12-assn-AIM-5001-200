## XML

#### Creating XML file
* For this section I learned from the reading and the XML Introduction article as a source to understand the XML format needed. I creates a table called books and then assign the tags of the columns to the text like the example below: 
    < Title > The Death and Life of Great American Cities < /Title >
#### Reading XML file
* I used urllib request library to retrieve the text and headers. The next step is to parse through the file and create the root variable. Using a for loop we append all the records and create the tag names for the data frame. Finally, use pd.DataFrame to convert into a data frame.

```python
#use urlretrieve to readthe xml file
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/AlejandraZS10/W12-assn-AIM-5001-200/master/books.xml')
#parses through the text content
parsed = objectify.parse(open(path))
#root variable by using get.root()
root = parsed.getroot()
```

```python
#empty list for our xml file content
pd2 = []
#iterates using the root of the file
for elt in root.book:
     # for each record, create a dict of tag names (e.g., 'YTD_ACTUAL')
    el_pd2 = {}
    for child in elt.getchildren():
        el_pd2[child.tag] = child.pyval
    #adds all records to the empty list
    pd2.append(el_pd2)
```

```python
# now check the results using pandas
books_xml = pd.DataFrame(pd2)

books_xml
```
